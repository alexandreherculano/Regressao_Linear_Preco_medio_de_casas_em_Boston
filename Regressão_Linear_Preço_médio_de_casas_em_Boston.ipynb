```
#importação de bibliotecas a criação de dataset inicial
from sklearn.datasets import load_boston
boston_dataset = load_boston()
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston['MEDV'] = boston_dataset.target
```

```
#análise exploratória - top 5
boston.head()
```

```
#análise exploratória - sumário estatístico
boston.describe()
```

```
#análise exploratória - correlação
boston.corr()
```

```
#plotagem - variáveis correlatas
plt.style.use('ggplot')
fig = boston.plot(kind='scatter', x='RM', y='MEDV', figsize=(8,4))


#plotagem - variáveis correlatas
plt.style.use('ggplot')
boston.plot(kind='scatter', x='LSTAT', y='MEDV', figsize=(8,4))

#plotagem - variáveis correlatas
plt.style.use('ggplot')
boston.plot(kind='scatter', x='PTRATIO', y='MEDV', figsize=(8,4))
```

```
#definição de variáveis de entrada
X = boston[['RM','LSTAT','PTRATIO']]
#definição de alvo
Y = boston['MEDV']
#separação de datset em amostra e treinamento
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test =  train_test_split(X, Y, test_size = 0.3, random_state=1)
#definição de modelo
from sklearn.linear_model import LinearRegression
model = LinearRegression()
#ajustar modelo
model.fit(X_train, Y_train)
```

```
#avalição de modelo - r_square
model.score(X_test, Y_test)
```
